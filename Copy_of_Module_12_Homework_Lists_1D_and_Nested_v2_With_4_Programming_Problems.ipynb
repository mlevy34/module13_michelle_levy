````python
# TODO: Implement F1 here
def add_item(items, item):
  items.append(item)
def remove_item(items, item):
  try:
    items.remove(item)
  except ValueError:
    print("item not in list")
def print_items(items):
  for idx, ch in enumerate(items, start = 1):
    print(idx, ch)

def save_items(items, filename):
  with open(filename, "w") as f:
    for line in items:
      f.write(line + "\n")
def load_items(filename):
  lst = []
  try:
    with open(filename, "r") as f:
      for line in f:
        lst.append(line)
      return lst
  except FileNotFoundError:
    print("File not found")

items = []
while True:
  print("\nMenu:")
  print("1) Add item")
  print("2) Remove item")
  print("3) View list")
  print("4) Save list")
  print("5) Load list")
  print("6) Quit")

  user_input = input("Choose an option number:")
  if user_input == "1":
    addend = input("Enter item you want to add:")
    add_item(items, addend)
  elif user_input == "2":
    removed_item = input("Enter item you want to remove:")
    remove_item(items, removed_item)
  elif user_input == "3":
    print_items(items)
  elif user_input == "4":
    filename = input("Enter filename:")
    save_items(items, filename)
  elif user_input == "5":
    filename= input("Enter filename:")
    load_items(filename)
  elif user_input == "6":
    break
  else:
    print("invalid choice")






````

### F3. Word Cleaner (file → 1D list → processing → file)
Read a text file (e.g., `paragraph.txt`) and build a 1D list of cleaned words.

**Requirements:**
- Read the file as text
- Convert to lowercase
- Remove punctuation characters like `. , ! ? ; :` (you may replace them with spaces)
- Split into words and store in a list `words`
- Build a list `long_words` containing only words with length ≥ 6
- Write `long_words` to `long_words.txt`, one per line

**If you don't have a `paragraph.txt`, create one in code first** (write 2–4 sentences).


````python
# TODO: Implement F3 here
try:
  with open("paragragh.txt", "w") as f:
    f.write(
        "Today\nwas\na\nvery\nhectic\nday,\nI\nhad\nso\nmuch\nto\ndo.\n"
         "Therefore,\nI\nam\nvery\ntired\nand\nwant\nto\ngo\nto\nsleep!")
except FileNotFoundError as ff:
  print(ff)
lst = []
try:
  with open("paragragh.txt", "r") as f:
    for line in f:
      lst.append(line)
      print(line)
except FileNotFoundError as ff:
  print(ff)
cleaned_list = []
for line in lst:
  new_line = line.lower().replace(".", " ").replace(",", " ").replace("!", " ")
  cleaned_list.append(new_line)
words = " ".join(cleaned_list)
cleaned_text = words.split()
long_words = []
for word in cleaned_text:
  if len(word) >= 6:
    long_words.append(word)
try:
  with open("long_words.txt", "w") as f:
    for line in long_words:
      f.write(line + "\n")
except FileNotFoundError as ff:
  print(ff)




````

### F4. Seating Chart (nested list) (2D list + loops + functions)
Create a seating chart using a nested list (list of rows).

**Requirements:**
- Build a 4×5 chart filled with the string `"EMPTY"`
- Write a function `assign_seat(chart, row, col, name)` that places a name
- Write a function `print_chart(chart)` that prints each row neatly
- Assign at least 6 names to different seats
- Count how many seats are still `"EMPTY"` and print that number

**Optional extension:** Write the seating chart to a file (one row per line).


````python
# TODO: Implement F4 here
row = 4
col = 5
count = 0
empty_seats = 0
# grid = [["EMPTY" for c in range(col)]for r in range(row)]
grid = []
for r in range(row):
  current_row = []
  for c in range(col):
    current_row.append("EMPTY")
  grid.append(current_row)
def assign_seat(chart, row, col, name):
  global count
  chart[row][col] = name
  count += 1
def print_chart(chart):
  for row in chart:
    print(row)
assign_seat(grid, 1, 2, 'Michelle')
assign_seat(grid, 0, 3, 'Max')
assign_seat(grid, 0, 0, 'Linda')
assign_seat(grid, 2, 2, 'Bella')
assign_seat(grid, 3, 2, 'Sarah')
assign_seat(grid, 1, 1, 'Sally')

empty_seats = (row * col) - count
print(f" Seats left: {empty_seats}")
print_chart(grid)


````
